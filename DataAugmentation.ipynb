```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img #gerekli fonksiyon
```

```python
datagen = ImageDataGenerator(rotation_range = 40,                            #yapilmak istenen islemler
                            width_shift_range = 0.2,
                            height_shift_range = 0.2,
                            shear_range = 0.2,
                            zoom_range = 0.2,
                            horizontal_flip = True,
                            vertical_flip = True, 
                            fill_mode = 'nearest')
```

```python
image = load_img('C:/Users/muham/PYTHON_ALISTIRMALARIM/veri artirma/BEN.PNG')             #hangi fografa
x = img_to_array(image)
x = x.reshape((1,)  + x.shape  )
```

```python
i = 0

for batch in datagen.flow(x, batch_size=1,
                          save_to_dir='C:/Users/muham/PYTHON_ALISTIRMALARIM/veri artirma/',     #nereye ve kac tane
                          save_format='jpg'):
    i += 1
    if i > 50:
        break 
```
